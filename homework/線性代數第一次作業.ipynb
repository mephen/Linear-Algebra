```python
#圓形公式： x**2 + y**2 = r**2 (原點=0的情況)
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

r = 1
a, b = (0., 0.)

x = np.linspace(a-r, a+r, 1000)
y = np.sqrt(r**2 - (x - a)**2)
plt.plot(x,y,'black')
plt.plot(x,-y,'black')
plt.axis('equal')

c, d = (2.6, 0)
x = np.linspace(c-r, c+r, 1000)
y = np.sqrt(r**2 - (x - c)**2)
plt.plot(x,y,'r')
plt.plot(x,-y,'r')

e, f = (-2.6, 0)
x = np.linspace(e-r, e+r, 1000)
y = np.sqrt(r**2 - (x - e)**2)
plt.plot(x,y,'b')
plt.plot(x,-y,'b')

g, h = (-1.3, -1.1)
x = np.linspace(g-r, g+r, 1000)
y = h + np.sqrt(r**2 - (x - g)**2)
plt.plot(x,y,'y')
plt.plot(x,-y-2.2,'y')

i, j = (1.3, -1.1)
x = np.linspace(i-r, i+r, 1000)
y = j + np.sqrt(r**2 - (x - i)**2)
plt.plot(x,y,'g')
plt.plot(x,-y-2.2,'g')
```
