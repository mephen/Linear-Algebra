<font size=5>1. 定義二維平面上的區域area</font>
<br>
<br>

```python
a, b = -2, 2
c, d = -2, 2

n = 20
x = np.linspace(a, b, n)
y = np.linspace(c, d, n)
```

```python
#另一個名area的list
area = []
for yj in y:
    for xi in x:
        plt.scatter(xi, yj, s=20, c='g') 
        area.append([xi, yj])

plt.xlim(-3, 3)
plt.ylim(-3, 3)

#轉置成方便向量乘法的column形式
area = np.array(area).T
area.shape
plt.title('area')
```

<br>
<font size=5>2. 定義線性轉換$\Phi$和A</font>
<br>
<br>

```python
tran = np.array([[math.cos(1/4*math.pi), -math.sin(1/4*math.pi)],
              [math.sin(1/4*math.pi), math.cos(1/4*math.pi)]])

def phi(x):
    assert tran.shape[1]==x.shape[0], 'incompatible dimension'
    return tran@x

#旋轉45度
tran21 = np.array([[math.cos(1/4*math.pi), -math.sin(1/4*math.pi)],
              [math.sin(1/4*math.pi), math.cos(1/4*math.pi)]])
#X延長8倍
tran22 = np.array([[8,0],
             [0,-1]])

def A(x):
    assert tran22.shape[1]==x.shape[0], 'incompatible dimension'
    return tran21@tran22@x
```
